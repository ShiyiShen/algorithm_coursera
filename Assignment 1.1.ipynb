```python
def prod(x,y):
    l = len(x)
    if l == 2:
        F = x[0] * y[0]
        G = x[1] * y[1]
        K = (x[0] * y[1]) + (x[1] * y[0])
        return F * 10**2 + K * 10 + G
    else:
        mid = l//2
        x_0 = x[:mid]
        x_1 = x[mid:]
        y_0 = y[:mid]
        y_1 = y[mid:]
        
        F = prod(x_0, y_0)
        G = prod(x_1, y_1)
        K = prod(x_0, y_1) + prod(x_1, y_0)
        return F * pow(10, l) + K * pow(10, l/2) + G
```
